```python
import matplotlib.pyplot as plt

copasBrasil = {
    1958 : "Suécia",
    1962 : "Chile",
    1970 : "México",
    1994 : "Estados Unidos",
    2002 : "Coreia/Japão",
}

anos = copasBrasil.keys()
pais = copasBrasil.values()


plt.figure(figsize=(8,4), facecolor="#FDDD00")
plt.title("Anos em que o Brasil foi campeão da copa", color='#012169')
plt.plot(pais, anos, color='#009739') # Gráfico de linha, país = x, anos = y
plt.xlabel('País sede', color = '#012169')
plt.ylabel('Ano',color = '#012169')
plt.show()
```

```python
# install openpyxl
import pandas as pd

df = pd.read_excel("Base_vendas.xlsx")

display(df.head())

df['Faturamento']= df["Qtd Vendida"] * df["Preço Unitário"]
```

```python
faturamentoLoja = df.groupby('Loja')['Faturamento'].sum().sort_values()

#display(faturamentoLoja) 

plt.figure(figsize=(15,4))
faturamentoLoja.plot(kind='barh')

plt.show()
```

```python
# fAZER um grafico de pizza mostrando a quantidade de produtos vendido por genero H/M

faturamentoPorGenero = df.groupby('Gênero')['Qtd Vendida'].sum()

#display(faturamentoLoja) 

plt.figure(figsize=(15,4))
pizza = faturamentoPorGenero.plot(kind='pie', colors= ['red', 'blue'], autopct='%1.1f%%',)
pizza.legend()
```
